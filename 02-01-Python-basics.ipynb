#### File Positions
- The tell() method tells you the current position within the file; in other words, the next read or write will occur at that many bytes from the beginning of the file.

- The seek(offset[, from]) method changes the current file position. The offset argument indicates the number of bytes to be moved. The from argument specifies the reference position from where the bytes are to be moved.

- If **from** is set to 0, it means use the beginning of the file as the reference position and 1 means use the current position as the reference position and if it is set to 2 then the end of the file would be taken as the reference position.
- If **from** is set to 2 (end of file) and if the file is opened in `text` mode, then the offset cannot be nonzero (it has to be zero)

```python
f = open('students53.abc', 'r')
f.seek(10)
print(f.read(1))
print(f.tell())
f.close()
```

#### Get the prices dictionary with LIVE prices rather than hardcoded or read from a file

```python
# prices = {'infy':1450, 'tcs':3200, 'lntinfo':4600, 'wipro':2800, 'hcltech':1200}
def getPrices(stockList, mydate):
    prices = {}
    for stock in stockList:
        df = yfinance.download(stock, mydate)
        price = df.loc[mydate]['Close']
        prices[stock] = round(price,2)
    return prices
```

```python
stockList = ['INFY.NS', 'TCS.NS', 'WIPRO.NS']
mydate = '2022-10-20'
prices = getPrices(stockList, mydate)
print(prices)
```

## Lambda Functions
- A lambda function is a **small anonymous** function
- It can have any number of parameters, but it can have only one expression
- lambda p1,p2,..pn : expression with p1,p2...pn

```python
def sum(x, y):
  result = x+y
  return result

sum(23, 54)
```

```python
myfunc = lambda x, y: x+y
myfunc(23,54)
```

```python
def convTemp(tempInF):
  tempInC = (5/9)*(tempInF - 32)
  return round(tempInC,2)

convTemp(85)
```

```python
convTempNew = lambda tempInF: round((5/9)*(tempInF - 32), 2)
convTempNew(85)
```

```python
class BankAccount():
  #Constructor
  def __init__(self, accNo, amount):
    self.bal = amount
    self.accNo = accNo
    print('Created account',self.accNo,'with bal',self.bal)
    return
  def deposit(self, amount):
    self.bal = self.bal + amount
    print('Balance for', self.accNo, 'is', self.bal)
    return
  def withdraw(self, amount):
    self.bal = self.bal - amount
    self.getBalance()
    return
  def getBalance(self):
    print('Balance for', self.accNo, 'is', self.bal)
    return
```
