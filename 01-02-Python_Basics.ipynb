```python
prices = {'infy':1500, 'tcs':3200, 'lntinfo' : 4500}
portfolio = {'infy':100, 'tcs': 200, 'lntinfo': 50}
portfolioValue = 0

minshares = 10
fundsAvailable = 200000
print('LnT info shares in portfoli:', portfolio['lntinfo'])
print('-----------------------------------------')
counter = 1
while (minshares * prices['lntinfo'] <= fundsAvailable):
  print("Loop counter:", counter)
  counter = counter + 1
  print('Buy',minshares,'shares of LnT Info')
  fundsAvailable = fundsAvailable - minshares * prices['lntinfo']
  portfolio['lntinfo'] = portfolio['lntinfo'] + minshares
  print('Funds available:', fundsAvailable)

print('-----------------------------------')
print('LnT info shares in portfoli:', portfolio['lntinfo'])
```

```python
def calcPortfolioValue(pf, pr):
  pv = 0
  for script in pf.keys():
    print(script, pf[script], pr[script])
    pv = pv + (pf[script] * pr[script])
  return pv
```
