```python
T1 = (1,2,3)
print(T1)
T2 = (1,)
# creating a new tuple which is a concatenation of tuple T1 and tuple T2 and assigning it to tuple T1
T1 = T1+T2
print(T2)
print(T1)
```

```python
# Concatenation, adding, appending -- are they different???
# All these are DIFFERENT OPERATIONS

# +  is an addition operation (when used for numbers)
# +, when used for lists, tuples and strings, works as a concatenation operator

T1 = (1,2,3)
print(T1)
T2 = (1,)
# creating a new tuple which is a concatenation of tuple T1 and tuple T2 and assigning it to tuple T1
T1 = T1+T2
print(T1)
s1 = 'Ram'
s2 = 'Kishore'
s3 = 'Prashanth'
team = s1 + ',' + s2 + ',' + s3
print(team)
```

- **+ symbol when used with numbers works as an addition operation**
- **+ symbol when used for lists, tuples and strings, works as a concatenation operator**

This is called a 'operator overloading' in OOPS parlance

### Sets (non-mutable)
- Lists and Tuples have inherent order
- Sets don't have inherent order

```python
s1 = {'apples','oranges','bananas'}
s2 = {'grapes','apples','pineapples'}
print(s1, type(s1))
print(s2, type(s2))
```

```python
s3 = s1.union(s2)
print(s3, type(s3))
```

```python
l1 = list(s1)
l2 = list(s2)
print(l1)
print(l2)
```

```python
l3 = l1 + l2
print(l3)
```

```python
s3 = s1.union(s2)
print(s3, type(s3))
s4 = s1.intersection(s2)
print(s4, type(s4))
```

```python
# sets - order not inherent
# sets - union
# sets - intersection
```
