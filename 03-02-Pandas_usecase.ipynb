### Putting it all together

```python
def email_gen(name):
    spname = name.split()
    emailid = '.'.join(spname) + '@edureka.co'
    return emailid.lower()

def pass_gen():
    # 1 to 3 uppercase characters
    caps_alpha = random.sample('ABCDEFGHIJKLMNOPQRSTUVWXYZ', random.randint(1,3))
    # 2 to 5 smallcase characters
    small_alpha = random.sample('abcdefghijklmnopqrstuvwxyz', random.randint(2,5))
    # 1 to 3 digits/numbers
    digits = random.sample('0123456789', random.randint(1,3))
    # 1 special character
    sp_chr = random.sample('#&$%@?!', 1)
    # All these should be concatenated and SHUFFLED
    pwdlist = caps_alpha + small_alpha + digits + sp_chr
    np.random.shuffle(pwdlist)
    pwd = ''.join(pwdlist)
    return pwd
```

```python
f = open('/content/names.txt')
allnames = f.read()
f.close()
namesList = allnames.split('\n')
namesList.pop()

emailList = [email_gen(name) for name in namesList]
pwdList = [pass_gen() for i in range(1000)]
empIdList = [id for id in range(2020001, 2020001+1000)]
print(namesList)
print(len(namesList))
print(emailList)
print(len(emailList))
print(pwdList)
print(len(pwdList))
print(empIdList)
print(len(empIdList))
```
